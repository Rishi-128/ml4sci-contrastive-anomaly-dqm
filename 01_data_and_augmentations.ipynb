# Contrastive Learning for Detector Anomaly Detection

In this project, we explore self-supervised contrastive learning as an alternative to reconstruction-based methods for anomaly detection in detector data quality monitoring (DQM).

The idea is to learn meaningful latent representations of detector behavior without relying on labeled anomalies. These representations can then be used to identify abnormal detector states based on their distance from normal patterns in embedding space.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Simulate detector data (normal behavior)
normal_data = np.random.normal(loc=0.0, scale=1.0, size=(10000, 10))

# Simulate anomalies
anomalies = np.random.normal(loc=3.0, scale=1.5, size=(500, 10))

data = np.vstack([normal_data, anomalies])
labels = np.hstack([np.zeros(len(normal_data)), np.ones(len(anomalies))])

print("Data shape:", data.shape)
```

```python
def augment_noise(x, noise_level=0.1):
    return x + noise_level * np.random.randn(*x.shape)

def augment_scaling(x, scale_range=(0.9, 1.1)):
    scale = np.random.uniform(*scale_range)
    return x * scale
```

```python
x = normal_data[:100]
x1 = augment_noise(x)
x2 = augment_scaling(x)

plt.figure(figsize=(6,4))
plt.plot(x[0], label="Original")
plt.plot(x1[0], label="Noise Augmented")
plt.plot(x2[0], label="Scaled")
plt.legend()
plt.title("Detector Data Augmentations")
plt.show()
```
